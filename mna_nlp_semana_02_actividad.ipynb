*   **Pregunta 1.**

Busca y elimina todos los saltos de línea '\n' que se encuentran al final de cada comentario.

*   **Pregunta 2.**  

Busca e imprime todas las palabras que terminan con dos o más signos de admiración seguidos, por ejemplo "!!!".

Debes imprimir tanto la palabra como la totalidad de signos de admiración que le siguen.

Indica cuántos resultados obtuviste.



```python
# Unimos todos los comentarios
texto = ' '.join(clean_comments)

# Busqueda de palabras que terminan con dos o mas signos de admiracion seguidos
resultados = re.findall(r'(\b\w+)(!{2,})', texto)

for p, s in resultados:
  print(f"Palabra: '{p}' - Cantidad de signos: '{len(s)}'")
print(f'Se encontraron {len(resultados)} resultados')
```

*   **Pregunta 3.**  

Busca e imprime todas las palabras que están escritas totalmente en mayúsculas. Cada coincidencia debe ser una sola palabra.

Indica cuántas palabras encontraste.



```python
# Busqueda de palabras que estén totalmente en mayúsculas
upper_words = re.findall(r'\b[A-Z]+\b', texto)

# Imprimimos los resultados
for p in upper_words:
    print(f"Palabra en mayúsculas: '{p}'")
```

```python
# Imprimimos el número total de palabras encontradas
print(f"Se encontraron {len(upper_words)} palabras totalmente en mayúsculas.")
```

*   **Pregunta 4.**  

Busca e imprime los comentarios en donde todos los caracteres alfabéticos (letras) están en mayúsculas.

Cada coincidencia encontrada debe ser todo el comentario/enunciado.

Indica cuántos resultados obtuviste.


```python
lineas = texto.splitlines()

# filtro para obtener caracteres alfabéticos en mayúsculas
upper_comments = [linea for linea in lineas if re.fullmatch(r'[^a-záéíóúñ]*', linea)]

for comentario in upper_comments:
    print(f"Comentario en mayúsculas: '{comentario}'")
```

```python

# Imprimimos cuántos resultados encontramos
print(f"Se encontraron {len(upper_comments)} comentarios totalmente en mayúsculas.")
```

*   **Pregunta 5.**  

Busca e imprime todas las palabras que tengan una vocal acentuada, del tipo á, é, í, ó, ú.

Indica cuántos resultados obtuviste.

```python
# Búsqueda de palabras que contengan al menos una vocal acentuada
p_acentuadas = re.findall(r'\b\w*[áéíóúÁÉÍÓÚ]\w*\b', texto)

for p in p_acentuadas:
    print(f"Palabra con acento: '{p}'")
```

*   **Pregunta 6.**

*   **Pregunta 7.**  

Busca e imprime todas las palabras que sean variantes de la palabra "love", sin importar si incluyen mayúsculas o minúsculas, o la manera en que esté conjugada o alguna otra variación que se haga con dicha palabra.

Indica cuántos resultados obtuviste.

```python
# Búsqueda, palabras que contengan "love" en cualquier forma, ignorando mayúsculas y minúsculas
variantes_love = re.findall(r'\b\w*love\w*\b', texto, flags=re.IGNORECASE)
for palabra in variantes_love:
    print(f"Palabra encontrada: '{palabra}'")
```

```python
print(f"Se encontraron {len(variantes_love)} palabras relacionadas con 'love'.")
```

*   **Pregunta 8.**  

Busca e imprime todas las palabras, variantes de "so" y "good", que tengan dos o más "o" en "so" y 3 o más "o" en good.

Indica cuántas encontraste.


```python
# Búsqueda de variantes de "so" con 2 o más letras "o"
variantes_so = re.findall(r'\bs[oO]{2,}\b', texto, flags=re.IGNORECASE)

# Búsqueda de variantes de "good" con 3 o más letras "o"
variantes_good = re.findall(r'\bg[oO]{3,}d\b', texto, flags=re.IGNORECASE)

# Juntamos todas las coincidencias
variantes_so_good = variantes_so + variantes_good

# Imprimimos los resultados
for palabra in variantes_so_good:
    print(f"Palabra encontrada: '{palabra}'")
```

```python
print(f"Se encontraron {len(variantes_so_good)} palabras variantes de 'so' o 'good'.")
```

*   **Pregunta 10.**  

Busca e imprime todas las palabras que inician con una letra mayúscula y terminan con una minúscula, pero que además no sea la primera palabra del comentario/string.

Indica la cantidad de resultados obtenidos.

```python
# - Inician con mayúscula [A-Z]
# - Siguen con cero o más caracteres alfabéticos
# - Terminan en minúscula [a-z]
# - NO son la primera palabra de la oración

# Separamos comentarios individuales
lineas = texto.splitlines()

palabras_encontradas = []

for linea in lineas:
    palabras = linea.split()
    # palabras = re.findall(r'\b\w+\b', linea)  #sin puntuacion
    for palabra in palabras[1:]:
        if  re.fullmatch(r'^[A-Z][a-z]+$', palabra):# re.fullmatch(r'[A-Z][a-zA-Z]*[a-z]', palabra):
            palabras_encontradas.append(palabra)

for palabra in palabras_encontradas:
    print(f"Palabra encontrada: '{palabra}'")
```

```python
print(f"Se encontraron {len(palabras_encontradas)} palabras.")
```

*   **Pregunta 11.**  

Busca e imprime la secuencia de dos o más palabras que están separadas por un guion, "-", sin que tengan espacios en blanco entre ellas.

Por ejemplo "Go-Kart" sería válido, pero "Go  -Kart" o "Go  -  Kart" no lo serían.

Indica la cantidad de resultados obtenidos.

*   **Pregunta 12.**  

Busca e imprime todas las palabras que terminan en "ing" o "ed".

Indica la cantidad de palabras que encontraste de cada una.

```python
# Busqueda de palabras que terminan en "ing" y busqueda de las que terminan en ed
palabras_ing = re.findall(r'\b\w+ing\b', texto, flags=re.IGNORECASE)
palabras_ed = re.findall(r'\b\w+ed\b', texto, flags=re.IGNORECASE)

print("Palabras que terminan en 'ing':")
for palabra in palabras_ing:
    print(f"'{palabra}'")

print("\nPalabras que terminan en 'ed':")
for palabra in palabras_ed:
    print(f"'{palabra}'")
```

#**Parte 3. Proceso de limpieza.**

*   **Pregunta 13.**  

Ahora realiza un proceso de limpieza del corpus que incluya los siguientes procesos:

*   Solo se deben considerar caracteres alfabéticos. Es decir, se eliminan todos los signos de puntuación y caracteres especiales.
*   Todos los caracteres alfabéticos se transforman a minúsculas.
*   Se deben eliminar todos los espacios en blanco adicionales que se puedan encontrar en cada comentario.

Al finalizar dicho proceso de limpieza, imprime el resultado de los primeros 10 comentarios resultantes.
   




```python
# lista para los comentarios limpios
cleaned_comments = []

# Iteramos sobre cada comentario original
for comment in clean_comments:
    # * Eliminamos todo lo que no sea alfabético (excepto espacios para separar palabras)
    comment = re.sub(r'[^a-zA-Z\s]', '', comment)
    # * Convertir todo a minúsculas
    comment = comment.lower()
    # * Eliminamos espacios en blanco adicionales (más de uno seguido → uno solo)
    comment = re.sub(r'\s+', ' ', comment).strip()
    cleaned_comments.append(comment)
```

```python
print("Primeros 10 comentarios limpios:\n")
for comment in cleaned_comments[:10]:
    print(f"- {comment}")
```
